```python
# Import required libraries
import pandas as pd
import plotly.express as px
import plotly.io as pio

# Load Data
df = pd.read_csv("./data/lightcast_job_postings.csv")

# 1. Salary Distribution by Employment Type
# Filter out records where salary is missing or zero
filtered_df = df[df['SALARY_FROM'] > 0]

# Create box plot
fig1 = px.box(
    filtered_df,
    x="EMPLOYMENT_TYPE_NAME",
    y="SALARY_FROM",
    title="Salary Distribution by Employment Type",
    color="EMPLOYMENT_TYPE_NAME",
    template="plotly_white"
)

# Customize layout
fig1.update_layout(
    title_x=0.5,
    title_font_size=20,
    xaxis_title="Employment Type",
    yaxis_title="Salary",
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    showlegend=False,
    height=600,
    width=1000
)

# Save the figure
fig1.write_html("_output/salary_by_employment.html")
print("Saved salary by employment visualization to _output/salary_by_employment.html")

# 2. Salary Distribution by Industry
# Create box plot
fig2 = px.box(
    filtered_df,
    x="NAICS2_NAME",
    y="SALARY_FROM",
    title="Salary Distribution by Industry",
    color="NAICS2_NAME",
    template="plotly_white"
)

# Customize layout
fig2.update_layout(
    title_x=0.5,
    title_font_size=20,
    xaxis_title="Industry",
    yaxis_title="Salary",
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    showlegend=False,
    height=600,
    width=1000,
    xaxis_tickangle=45  # Rotate x-axis labels for better readability
)

# Save the figure
fig2.write_html("_output/salary_by_industry.html")
print("Saved salary by industry visualization to _output/salary_by_industry.html")

# 3. Job Posting Trends Over Time
# Convert POSTED to datetime and count posts per date
df['POSTED'] = pd.to_datetime(df['POSTED'])
posting_trends = df.groupby('POSTED').size().reset_index(name='count')

# Create line chart
fig3 = px.line(
    posting_trends,
    x="POSTED",
    y="count",
    title="Job Posting Trends Over Time",
    template="plotly_white"
)

# Customize layout
fig3.update_layout(
    title_x=0.5,
    title_font_size=20,
    xaxis_title="Date",
    yaxis_title="Number of Job Postings",
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    height=600,
    width=1000
)

# Save the figure
fig3.write_html("_output/posting_trends.html")
print("Saved posting trends visualization to _output/posting_trends.html")
```

```python
# 4. Top 10 Job Titles by Count
# Count occurrences of each job title
job_counts = df['TITLE_NAME'].value_counts().head(10).reset_index()
job_counts.columns = ['TITLE_NAME', 'count']

# Create bar chart
fig4 = px.bar(
    job_counts,
    x="TITLE_NAME",
    y="count",
    title="Top 10 Job Titles by Count",
    color="count",
    template="plotly_white",
    color_continuous_scale="Viridis"  # Using a professional color scale
)

# Customize layout
fig4.update_layout(
    title_x=0.5,
    title_font_size=20,
    xaxis_title="Job Title",
    yaxis_title="Number of Postings",
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    height=600,
    width=1000,
    xaxis_tickangle=45,  # Rotate x-axis labels for better readability
    font=dict(family="Arial", size=12),  # Set font family and size
    plot_bgcolor="white",  # Set background color
    paper_bgcolor="white"  # Set paper background color
)

# Save the figure
fig4.write_html("_output/top_job_titles.html")
print("Saved top job titles visualization to _output/top_job_titles.html")

# Add explanation
print("\nExplanation:")
print("1. This visualization shows the most frequently posted job titles in the dataset.")
print("2. The height of each bar represents the number of job postings for that specific title.")
```

```python
# 5. Remote vs On-Site Job Postings
# Count job postings by remote type
remote_counts = df['REMOTE_TYPE_NAME'].value_counts().reset_index()
remote_counts.columns = ['REMOTE_TYPE_NAME', 'count']

# Create pie chart
fig5 = px.pie(
    remote_counts,
    values="count",
    names="REMOTE_TYPE_NAME",
    title="Remote vs On-Site Job Postings",
    template="plotly_white",
    color_discrete_sequence=px.colors.qualitative.Set3,  # Using a professional color palette
    hole=0.4  # Create a donut chart for better visualization
)

# Customize layout
fig5.update_layout(
    title_x=0.5,
    title_font_size=20,
    height=600,
    width=1000,
    font=dict(family="Arial", size=12),  # Set font family and size
    plot_bgcolor="white",  # Set background color
    paper_bgcolor="white",  # Set paper background color
    legend=dict(
        orientation="h",  # Horizontal legend
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

# Add percentage labels
fig5.update_traces(
    textposition='inside',
    textinfo='percent+label'
)

# Save the figure
fig5.write_html("_output/remote_vs_onsite.html")
print("Saved remote vs onsite visualization to _output/remote_vs_onsite.html")

# Add explanation
print("\nExplanation:")
print("1. This visualization shows the distribution of job postings between remote and on-site positions.")
print("2. The size of each slice represents the proportion of job postings for that work arrangement type.")
```

```python
# 6. Skill Demand Analysis by Industry
# First, let's get the top 5 skills
top_skills = df['SKILLS_NAME'].str.split(',').explode().str.strip().value_counts().head(5).index

# Create a new column for each top skill
for skill in top_skills:
    df[f'has_{skill}'] = df['SKILLS_NAME'].str.contains(skill, na=False)

# Group by industry and count skills
skill_columns = [f'has_{skill}' for skill in top_skills]
skill_by_industry = df.groupby('NAICS2_NAME')[skill_columns].sum().reset_index()

# Create stacked bar chart
fig6 = px.bar(
    skill_by_industry,
    x="NAICS2_NAME",
    y=skill_columns,
    title="Top Skills Demand by Industry",
    template="plotly_white",
    barmode="stack",
    color_discrete_sequence=px.colors.qualitative.Set3  # Using a professional color palette
)

# Customize layout
fig6.update_layout(
    title_x=0.5,
    title_font_size=20,
    xaxis_title="Industry",
    yaxis_title="Number of Job Postings",
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    height=600,
    width=1000,
    xaxis_tickangle=45,  # Rotate x-axis labels for better readability
    font=dict(family="Arial", size=12),  # Set font family and size
    plot_bgcolor="white",  # Set background color
    paper_bgcolor="white",  # Set paper background color
    legend=dict(
        orientation="h",  # Horizontal legend
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

# Save the figure
fig6.write_html("_output/skill_demand.html")
print("Saved skill demand visualization to _output/skill_demand.html")

# Add explanation
print("\nExplanation:")
print("1. This visualization shows the distribution of top 5 most demanded skills across different industries.")
print("2. Each colored segment in the stacked bars represents the number of job postings requiring that specific skill in each industry.")
```

```python
# 7. Salary Analysis by ONET Occupation Type
# Calculate median salary and count for each ONET occupation
onet_analysis = df.groupby('ONET_NAME').agg({
    'SALARY_FROM': 'median',
    'ID': 'count'
}).reset_index()
onet_analysis.columns = ['ONET_NAME', 'median_salary', 'job_count']

# Create bubble chart
fig7 = px.scatter(
    onet_analysis,
    x="ONET_NAME",
    y="median_salary",
    size="job_count",
    title="Salary Analysis by ONET Occupation Type",
    template="plotly_white",
    color="job_count",
    color_continuous_scale="Viridis"  # Using a professional color scale
)

# Customize layout
fig7.update_layout(
    title_x=0.5,
    title_font_size=20,
    xaxis_title="ONET Occupation",
    yaxis_title="Median Salary",
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    height=600,
    width=1000,
    xaxis_tickangle=45,  # Rotate x-axis labels for better readability
    font=dict(family="Arial", size=12),  # Set font family and size
    plot_bgcolor="white",  # Set background color
    paper_bgcolor="white",  # Set paper background color
    coloraxis_colorbar=dict(
        title="Number of Jobs"
    )
)

# Save the figure
fig7.write_html("_output/onet_salary.html")
print("Saved ONET salary analysis visualization to _output/onet_salary.html")

# Add explanation
print("\nExplanation:")
print("1. This visualization shows the relationship between ONET occupation types, their median salaries, and the number of job postings.")
print("2. The size and color of each bubble represents the number of job postings, while the y-axis position shows the median salary for that occupation.")
```

```python
# 8. Career Pathway Trends (Alternative Visualization)
# Count transitions between SOC levels
soc_transitions = df.groupby(['SOC_2021_2_NAME', 'SOC_2021_3_NAME']).size().reset_index(name='count')

# Sort by count to get the most common transitions
soc_transitions = soc_transitions.sort_values('count', ascending=False).head(20)

# Create a grouped bar chart
fig8 = px.bar(
    soc_transitions,
    x="SOC_2021_2_NAME",
    y="count",
    color="SOC_2021_3_NAME",
    title="Career Pathway Trends",
    template="plotly_white",
    color_discrete_sequence=px.colors.qualitative.Set3  # Using a professional color palette
)

# Customize layout
fig8.update_layout(
    title_x=0.5,
    title_font_size=20,
    xaxis_title="Major Occupational Group",
    yaxis_title="Number of Job Postings",
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    height=800,
    width=1200,
    xaxis_tickangle=45,  # Rotate x-axis labels for better readability
    font=dict(family="Arial", size=12),  # Set font family and size
    plot_bgcolor="white",  # Set background color
    paper_bgcolor="white",  # Set paper background color
    legend=dict(
        orientation="h",  # Horizontal legend
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

# Save the figure
fig8.write_html("_output/career_pathways.html")
print("Saved career pathways visualization to _output/career_pathways.html")

# Add explanation
print("\nExplanation:")
print("1. This visualization shows the distribution of job postings across different occupational categories.")
print("2. Each bar represents a major occupational group, with different colors indicating the specific occupations within that group.")
```
